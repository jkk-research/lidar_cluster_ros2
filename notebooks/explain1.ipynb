```python
import numpy as np
import matplotlib.pyplot as plt

def angle_diff(a, b):
    while a < 0.0:
        a += np.pi * 2
    while a > np.pi * 2:
        a -= np.pi * 2        
    while b < 0.0:
        b += np.pi * 2
    while b > np.pi * 2:
        b -= np.pi * 2       
    if a > b:
        greater = a
        smaller = b
    else:
        greater = b
        smaller = a
    res = greater - smaller
    print("result: %.3f rad [%.1f deg]" % (res, np.rad2deg(res)))
    return res

def calcPointPos(range, rad_ang):
    x1 = range * np.cos(rad_ang)
    y1 = range * np.sin(rad_ang)
    return y1, x1 * -1

def plot_angle_bg():
    plt.figure(figsize=(6, 6))
    tmp = np.arange(0.0, 2 * np.pi, np.deg2rad(30), dtype=object)
    for t in tmp:
        tx, ty = calcPointPos(1, t)
        plt.plot([0, tx], [0, ty], 'y-')
        plt.text(tx*1.1, ty*1.1, "%.0f" % (np.rad2deg(t)), fontsize = 10, color='#aaa')

def plot_angle(rad_ang, name="unknown", scale=1.4):
    rx, ry = calcPointPos(1, rad_ang)
    plt.plot([0.0, rx], [0.0, ry], '*--', label=name)
    plt.text(rx*scale, ry*scale, "%.1f" % (np.rad2deg(rad_ang)), fontsize = 9)


# a = np.deg2rad(90.1)
# b = np.deg2rad(480.2)
a = np.random.rand(1) * np.pi * 4 - np.pi * 2
b = np.random.rand(1) * np.pi * 2

plot_angle_bg()
plot_angle(a, "a")
plot_angle(b, "b")
plot_angle(angle_diff(a, b), "a-b")
# plot_angle(angle_diff(b, a), "b-a", 1.5)
# plot_angle(angle_diff(a, 0), "a_", 1.2)
plt.axis('equal')
plt.xticks([])
plt.yticks([])
plt.grid()
plt.legend()
plt.axis('off')
plt.show()
```
